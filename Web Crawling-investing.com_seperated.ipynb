```python
from selenium import webdriver
import time
import datetime as dt
from dateutil.parser import parse
import pandas as pd
from datetime import datetime
from selenium.webdriver.common.keys import Keys
```

```python
dict_combine = {}
title_list = []
date_list =[]

#get to the page with the articles 
driver = webdriver.Chrome('./chromedriver')
page = 'https://www.investing.com/equities/royal-caribbean-news'
driver.get(page)

#parameter for the iteration
i = 1
web_page = 1

while True:
    print(web_page,'page')
    #get datetime data 
    try:
        try:
            date  = driver.find_element_by_xpath(f'//*[@id="leftColumn"]/div[8]/article[{i}]/div[1]/span/span[2]').text
   
        except:
            date  = driver.find_element_by_xpath(f'//*[@id="leftColumn"]/div[8]/article[{i}]/div[1]/div/span[2]').text
    
        try:
            date_datetime = parse(date).date()
            print(date_datetime)
   
        except:
            date_datetime = datetime.today().date()
            print(date_datetime)
    
        #if it has been more than 3 years, then stop crawling
        if date_datetime.year == 2018:
            break
        date_list.append(date_datetime)
        
        #get the title of the news
        title_string = driver.find_element_by_xpath(f'//*[@id="leftColumn"]/div[8]/article[{i}]/div[1]/a').text
        print(title_string)
        title_list.append(title_string)
        
    
        #get the article below
        i += 1
        if i % 11 == 0:
            if web_page >=  6:
                page_button = driver.find_element_by_xpath('//*[@id="paginationWrap"]/div[2]/a[6]')
            
            else:
                page_button = driver.find_element_by_xpath(f'//*[@id="paginationWrap"]/div[2]/a[{web_page+ 1}]')
        
            #go to the next page
            page_button.send_keys(Keys.ENTER)
            time.sleep(3)
            web_page += 1
            i = 1
                
    except:
        #when there are only 9 articles on page
        if web_page >=  6:
            page_button = driver.find_element_by_xpath('//*[@id="paginationWrap"]/div[2]/a[6]')
            
        else:
            page_button = driver.find_element_by_xpath(f'//*[@id="paginationWrap"]/div[2]/a[{web_page+ 1}]')
            
        #go to the next page
        page_button.send_keys(Keys.ENTER)
        time.sleep(3)
        web_page += 1
        i = 1
                
            
```

```python
df_investing_list = pd.DataFrame(title_list, index = date_list)
df_investing_list.head()
```
