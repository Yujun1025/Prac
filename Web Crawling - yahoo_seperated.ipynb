```python
from selenium import webdriver
import time
import datetime as dt
from dateutil.parser import parse
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver import ActionChains
import pandas as pd
from datetime import datetime
```

```python
dict_combine = {}
title_list = []
date_list =[]

#get to the page with the articles 
driver = webdriver.Chrome('./chromedriver')
page = 'https://finance.yahoo.com/quote/RCL?p=RCL&.tsrc=fin-srch'
driver.get(page)

#parameter for the iteration
i = 1

while True:
    try:

        try:
            #if it's an ad, then pass
            if driver.find_element_by_xpath(f'//*[@id="quoteNewsStream-0-Stream"]/ul/li[{i}]/div/div[1]/div/div/a[1]').text == ('Ad' or '광고'):
                print(f'{i}th article is an ad')
                pass
    
        except NoSuchElementException:   
            try:
                #Click the news title with picture
                para = driver.find_element_by_xpath(f'//*[@id="quoteNewsStream-0-Stream"]/ul/li[{i}]/div/div/div[2]/h3/a')
                para.click()   
                time.sleep(3)
            except NoSuchElementException:
                #Click the news title without picture
                para = driver.find_element_by_xpath(f'//*[@id="quoteNewsStream-0-Stream"]/ul/li[{i}]/div/div/div[1]/h3/a')
                para.click()   
                time.sleep(3)
            
            #get datetime data and if its been more than 3 years, then stop crawling
            date  = driver.find_element_by_class_name('caas-attr-time-style').text
            date_datetime = parse(date).date()
            print(date_datetime)
            if date_datetime.year == 2018:
                break
            date_list.append(date_datetime)
            
            #get the title of the news
            title_string = driver.find_element_by_class_name('caas-title-wrapper').text
            print(title_string)
            title_list.append(title_string)
            
            #go back to previous page
            driver.back()        
            
            # scroll the page down
            action = ActionChains(driver)
            action.move_to_element(para).perform()
            time.sleep(1)
        #get the article below
        i += 1
            
    #when there are no more articles to crwal, break
    except:
        print('There are no more articles')
        break
        
```

```python
df_yahoo_list = pd.DataFrame(title_list, index = date_list)
df_yahoo_list.head()
```
